```python
import pandas as pd
```

```python
premier_league = pd.read_excel('../project_data/premier_league_games_full.xlsx')
```

```python
premier_league.head()
```

```python
premier_league.groupby("HomeTeam")
```

```python
premier_league.groupby("HomeTeam")[['HomeGoals']].mean().sort_values('HomeGoals', ascending=False)
```

```python
premier_league.groupby("AwayTeam")[['AwayGoals']].mean().sort_values('AwayGoals', ascending=False)
```

```python
premier_league.groupby("AwayTeam")[['AwayGoals']].sum().sort_values('AwayGoals', ascending=False)
```
