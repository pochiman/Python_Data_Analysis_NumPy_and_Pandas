```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_csv("../retail/retail_2016_2017.csv", parse_dates=["date"])

retail.head()
```

```python
retail_agg = retail.groupby(['date']).agg({"sales": "sum"}).round().iloc[1:10]

retail_agg
```

```python
retail_agg = retail.groupby(['date']).agg({"sales": "sum"}).round().iloc[1:10]

retail_agg.assign(daily_change = retail_agg["sales"].diff())
```

```python
retail_agg = retail.groupby(['date']).agg({"sales": "sum"}).round().iloc[1:10]

retail_agg.assign(daily_change = (retail_agg["sales"].diff() / retail_agg["sales"].shift()))
```
