```python
import pandas as pd
```

```python
premier_league = pd.read_excel('../project_data/premier_league_games_full.xlsx')
```

```python
premier_league.head()
```

```python
premier_league.query("HomeTeam in ['Arsenal', 'Chelsea', 'Everton']").pivot_table(index="HomeTeam", 
                                                                                  columns="season", 
                                                                                  values="HomeGoals",
                                                                                  aggfunc="mean",
                                                                                  margins=True) # .style.background_gradient(cmap="RdYlGn", axis=None)
```

```python
premier_league.query("HomeTeam in ['Arsenal', 'Chelsea', 'Everton']").pivot_table(index="HomeTeam", 
                                                                                  columns="season", 
                                                                                  values="HomeGoals",
                                                                                  aggfunc="mean",
                                                                                  margins=True).style.background_gradient(cmap="RdYlGn", axis=None)
```
