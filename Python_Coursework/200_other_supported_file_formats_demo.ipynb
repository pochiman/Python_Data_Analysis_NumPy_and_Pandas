```python
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_GDP'

gdp_data = pd.read_html(url)[0]
```

```python
gdp_data.head()
```

```python
gdp_data = (gdp_data
            .droplevel(level=0, axis=1)
            .drop(index=0))
```

```python
gdp_data.head()
```

```python
(gdp_data
 .set_index('State or federal district')
 .iloc[:10, 0]
 .plot.bar());
```
