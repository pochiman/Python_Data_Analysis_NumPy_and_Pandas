```python
import pandas as pd
```

```python
premier_league = pd.read_excel('../project_data/premier_league_games_full.xlsx')
```

```python
premier_league.head()
```

```python
pm = premier_league.query("HomeTeam in ['Arsenal', 'Chelsea', 'Everton']").pivot_table(index="HomeTeam", 
                                                                                       columns="season", 
                                                                                       values="HomeGoals",
                                                                                       aggfunc="mean",)
```

```python
pm
```

```python
pm.melt(# id_vars="HomeTeam",
                      # value_vars=["2008/2009", "2009/2010", "2010/2011"],
                      # value_name='avg_goals'
)
```

```python
pm.reset_index().melt(# id_vars="HomeTeam",
                      # value_vars=["2008/2009", "2009/2010", "2010/2011"],
                      # value_name='avg_goals'
)
```

```python
pm.reset_index().melt(id_vars="HomeTeam",
                      # value_vars=["2008/2009", "2009/2010", "2010/2011"],
                      # value_name='avg_goals'
)
```

```python
pm.reset_index().melt(id_vars="HomeTeam",
                      value_vars=["2008/2009", "2009/2010", "2010/2011"],
                      # value_name='avg_goals'
)
```

```python
pm.reset_index().melt(id_vars="HomeTeam",
                      value_vars=["2008/2009", "2009/2010", "2010/2011"],
                      value_name='avg_goals'
)
```
