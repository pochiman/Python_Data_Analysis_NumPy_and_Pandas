```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_csv("../retail/retail_2016_2017.csv", parse_dates=["date"])

retail.head()
```

```python
retail.groupby([retail["date"].dt.year, retail["date"].dt.month]).agg("sum").round(2)
```

```python
retail.groupby([retail["date"].dt.year, retail["date"].dt.month]).sum(numeric_only=True).round(2)
```

```python
retail.set_index("date").resample("M").sum().round(2)
```

```python
retail.set_index("date").resample("Q").sum().round(2)
```

```python
retail.set_index("date").resample("Y").sum().round(2)
```
