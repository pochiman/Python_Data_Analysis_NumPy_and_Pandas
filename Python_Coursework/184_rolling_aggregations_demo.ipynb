```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_csv("../retail/retail_2016_2017.csv", parse_dates=["date"])

retail.head()
```

```python
retail.set_index("date").resample("Y").sum().round(2)
```

```python
retail_daily = retail.set_index("date").loc[:, "sales"].resample("D").sum().round(2)

retail_daily
```

```python
retail_daily.rolling(2).mean() # .plot()
```

```python
retail_daily.plot()
```

```python
retail_daily.rolling(2).mean().plot()
```

```python
retail_daily.rolling(30).mean().plot()
```
