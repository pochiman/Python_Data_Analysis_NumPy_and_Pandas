```python
import pandas as pd
```

```python
premier_league = pd.read_excel('../project_data/premier_league_games_full.xlsx')
```

```python
premier_league.head()
```

```python
(premier_league
 .groupby(["season", "HomeTeam"], as_index=False)
 .agg("sum"))
```

```python
(premier_league
 .groupby(["season", "HomeTeam"], as_index=False)
 .agg(["sum", "mean"]))
```

```python
(premier_league
 .groupby(["season", "HomeTeam"], as_index=False)
 .agg(["sum", "count"]))
```

```python
(premier_league
 .groupby(["season", "HomeTeam"], as_index=False)
 .agg({"HomeGoals": ["sum", "mean"],
       "AwayGoals": ["sum", "sum"]}))
```

```python
(premier_league
 .groupby(["season", "HomeTeam"], as_index=False)
 .agg(home_goal_sum = ("HomeGoals", "sum"),
      away_goal_sum = ("AwayGoals", "sum")))
```
